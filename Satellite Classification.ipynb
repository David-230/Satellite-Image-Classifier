```python
import os
import cv2
import numpy as np
from matplotlib import pyplot
from matplotlib import pyplot
from matplotlib.image import imread
from tensorflow.keras.preprocessing.image import img_to_array, load_img
```

```python
def visualize_images_by_class(parent_folder):
    # Get a list of all subfolders (each subfolder represents a class)
    class_folders = [f for f in os.listdir(parent_folder) if os.path.isdir(os.path.join(parent_folder, f))]
    
            
    for i, class_folder in enumerate(class_folders):
        class_path = os.path.join(parent_folder, class_folder)
        class_image_files = [f for f in os.listdir(class_path) if f.endswith(('.png', '.jpg', '.jpeg', '.gif'))]

        for f, filename in enumerate(class_image_files):
            if f < 4:
                filename2 = os.path.join(class_path, filename)
                image = imread(filename2)

                # Adjusting subplot parameters for a 2x2 grid
                pyplot.subplot(4, 4, 1 + f)
                pyplot.imshow(image)
                pyplot.title(f"Class: {class_folder}", fontsize=8)  # Optional: Display class name as title
                
        #pyplot.suptitle(f"Class: {class_folder}", fontsize=16)        
        pyplot.show()
    
parent_folder = r"C:\Users\David\Documents\Python projects\Image Classiffication\train"
visualize_images_by_class(parent_folder)
```

```python
#directory structure
data_dir = r"C:\Users\David\Documents\Python projects\Image Classiffication\train"
class_labels = ['cloudy', 'desert', 'green_area', 'water']

# Function to load and preprocess images
def load_and_preprocess_images(data_dir, class_labels):
    image_data = []
    labels = []

    for label in class_labels:
        label_path = os.path.join(data_dir, label)
        for img_file in os.listdir(label_path):
            img_path = os.path.join(label_path, img_file)
            img = cv2.imread(img_path)  #open images
            target_size=(224, 224) #resize to 224x224
            image = cv2.resize(img, target_size) #resize the images
            image_array = np.array(image) #convert the images to array
            #image_array = image_array.astype(np.float32) / 255.0 #normalize the images
            image_data.append(image_array)
            labels.append(label)

    return image_data, labels

# Load and preprocess images
image_data, labels = load_and_preprocess_images(data_dir, class_labels)
```
